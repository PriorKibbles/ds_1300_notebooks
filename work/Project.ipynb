```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#create a composite score for each team and list top 5 composites and highlight top team
df = pd.read_csv('../work/2010-2011.csv',delimiter=",")
df["Composite"] = (df["W"]/df["L"])*df["NRtg"]+df["MOV"]
df2011 = df.sort_values(by="Composite", ascending = False)
df2011.head(8)
```

```python
plt.rcdefaults()
fig, ax = plt.subplots()

# Example data
subset = df2011[df2011['Composite']>12.29]
y_pos = np.arange(len(subset))

ax.barh(y_pos, subset['Composite'], align='center',color=['silver','silver','silver','silver','silver','silver','royalblue','silver'])
ax.set_yticks(y_pos)
ax.set_yticklabels(subset.Team.tolist())
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Composite Score')
ax.set_title('2010-2011 Top Contenders Based on Composite')

plt.show()
```
